### 1. Encuentre las dimensiones del cilindro circular recto de volumen máximo que puede inscribirse en un cono circular recto dado



**Paso 1.** Definimos la función objetivo $Q$, que es el volumen del cilindro de radio $r$ y altura $h$:

$$
Q(r,h) \;=\;\pi\,r^2\,h.
$$



**Paso 2.** Usamos la geometría del cono de radio de base $R$ y altura $H$. Si apoyamos el cilindro sobre la base del cono, la sección transversal del cono a una altura $h$ sobre la base tiene radio

$$
r_{\text{cono}}(h)\;=\;R\Bigl(1 - \tfrac{h}{H}\Bigr).
$$

Para que el cilindro quepa justo dentro del cono, su radio debe coincidir con esa sección:

$$
r = R\Bigl(1 - \tfrac{h}{H}\Bigr).
$$

Sustituyendo en $Q$, queda como función de la sola variable $h$:

$$
Q(h)
= \pi \bigl[R(1 - \tfrac{h}{H})\bigr]^2\,h
= \pi R^2\,h\bigl(1 - \tfrac{h}{H}\bigr)^2.
$$


**Paso 3.** Buscamos los puntos críticos de $Q(h)$ en el intervalo $[0,H]$.

1. **Fronteras:**

   * $h=0$  $\Rightarrow Q(0)=0$.
   * $h=H$  $\Rightarrow Q(H)=0$.

2. **Puntos estacionarios:**

   $$
   Q(h)=\pi R^2\,h\,(1 - \tfrac{h}{H})^2.
   $$

   Derivamos respecto a $h$:

   $$
   Q'(h)
   = \pi R^2\Bigl[(1 - \tfrac{h}{H})^2 + h\cdot 2(1 - \tfrac{h}{H})\bigl(-\tfrac1H\bigr)\Bigr].
   $$

   Factorizamos:

   $$
   Q'(h)
   = \pi R^2\,(1 - \tfrac{h}{H})\Bigl[(1 - \tfrac{h}{H}) - \tfrac{2h}{H}\Bigr]
   $$

   $$
   = \pi R^2(1 - \tfrac{h}{H})(1 - \tfrac{3h}{H}).
   $$

   Igualamos a cero:

   $$
   (\pi R^2)(1 - \tfrac{h}{H})(1 - \tfrac{3h}{H}) = 0
   \quad\Longrightarrow\quad
   h = H
   \quad\text{o}\quad
   h = \tfrac{H}{3}.
   $$

3. **Punto singular:** no hay problemas de dominio interno aparte de $h\in[0,H]$.



**Paso 4.** Comprobamos cuál de estos puntos da el máximo:

* En $h=0$ y $h=H$, $Q=0$.
* En $h = \tfrac{H}{3}$,

  $$
  Q\bigl(\tfrac{H}{3}\bigr)
  = \pi R^2\;\tfrac{H}{3}\;\Bigl(1 - \tfrac{1}{3}\Bigr)^2
  = \pi R^2\;\tfrac{H}{3}\;\Bigl(\tfrac{2}{3}\Bigr)^2
  = \tfrac{4}{27}\,\pi R^2 H >0.
  $$

Además, la segunda derivada en $h=\tfrac H3$ resulta negativa, confirmando máximo.



**Conclusión.** El cilindro de volumen máximo inscrito tiene

$$
h = \frac{H}{3}
$$

$$ r = R\Bigl(1 - \tfrac{1}{3}\Bigr) = \frac{2R}{3} $$


### 2. Suponga que un pez nada río arriba con velocidad relativa al agua $v$ y que la corriente del río tiene velocidad $-v_c$ (el signo negativo indica que la velocidad de la corriente es en dirección opuesta a la del pez). La energía empleada en recorrer una distancia $d$ a contracorriente es directamente proporcional al tiempo requerido para recorrer la distancia $d$ y el cubo de la velocidad. ¿Qué velocidad $v$ minimiza la energía empleada en nadar esta distancia?



**Formulación de la función objetivo**

La energía $E$ empleada es proporcional al tiempo de nado $t$ y al cubo de la velocidad relativa al agua $v$. (k es la constante de proporcionalidad)

$$ E = k\,t\,v^3 $$
Dado que la corriente opone velocidad $v_c$, la velocidad neta río arriba es $v - v_c$ (con $v>v_c$). Para una distancia $d$,

$$
t = \frac{d}{v - v_c},
$$

luego

$$
E = k\,t\,v^3 = k\,\frac{d}{v - v_c}\,v^3,
$$

siendo $k>0$ constante. Para minimizar $E$ basta trabajar con

$$
Q(v) = \frac{v^3}{\,v - v_c\,},
\qquad v>v_c.
$$

**Cálculo de puntos críticos**

* **Fronteras del dominio**
  $\displaystyle \lim_{v\to v_c^+}Q(v)=+\infty,\quad \lim_{v\to+\infty}Q(v)=+\infty$.
* **Estacionarios**

  $$
    Q'(v)
    = \frac{(3v^2)(v - v_c) - v^3}{(v - v_c)^2}
    = \frac{v^2(2v - 3v_c)}{(v - v_c)^2}
    \;\Longrightarrow\; 2v - 3v_c = 0
    \;\Longrightarrow\; v = \tfrac{3}{2}v_c.
  $$
* **Singularidades**
  El único punto no derivable sería $v=v_c$, donde $Q$ no está definido, por lo que no compite como mínimo.

\

En $v=\tfrac32\,v_c$ la derivada cambia de signo de negativa a positiva y los valores en las fronteras son infinitos. Por tanto, el **mínimo absoluto** de $Q$ (y de la energía) se alcanza en

$$
\boxed{v = \frac{3}{2}\,v_c.}
$$


### 3. Un pasillo de 6 pies de ancho da vuelta en ángulo recto. ¿Cuál es la longitud de la varilla delgada más larga que puede transportarse alrededor de la esquina, suponiendo que la varilla no puede doblarse?


**Formulación de la función objetivo**

Sea $L$ la longitud de la varilla. Si gira en ángulo recto entre dos pasillos de anchura $6$ ft y forma un ángulo $\theta$ con la pared de uno de ellos, entonces la parte en el primer pasillo mide

$$
\frac{6}{\sin\theta},
$$

y la parte en el segundo pasillo mide

$$
\frac{6}{\cos\theta}.
$$

Por tanto,

$$
L(\theta)
= \frac{6}{\sin\theta} + \frac{6}{\cos\theta},
\qquad
0<\theta<\tfrac{\pi}{2}.
$$

**Puntos críticos**

* **Fronteras:**
  $\displaystyle \lim_{\theta\to0^+}L(\theta)=+\infty,\quad \lim_{\theta\to(\pi/2)^-}L(\theta)=+\infty.$
* **Estacionarios:**

  $$
    L'(\theta)
    =6\Bigl(-\frac{\cos\theta}{\sin^2\theta} + \frac{\sin\theta}{\cos^2\theta}\Bigr)
    =0

    \frac{\sin\theta}{\cos^2\theta}
    =\frac{\cos\theta}{\sin^2\theta}
    \;\Longrightarrow\;
    \tan^3\theta=1
    \;\Longrightarrow\;
    \theta=\frac{\pi}{4}.
  $$
* **Singularidades:**
  No hay derivada en las fronteras, pero allí $L\to\infty$, por lo que no compiten como mínimo.

\

En $\theta=\tfrac{\pi}{4}$ la segunda derivada es positiva y los valores en las fronteras son infinitos, por lo que $L$ alcanza su **mínimo absoluto** en

\


$$
L_{\min}
= L\Bigl(\tfrac{\pi}{4}\Bigr)
= \frac{6}{\sin(\pi/4)} + \frac{6}{\cos(\pi/4)}
= 6\sqrt2 + 6\sqrt2
= 12\sqrt2\text{ ft.}
$$


### 4. Un objeto que se arroja desde el borde de un acantilado de 100 pies, sigue la trayectoria dada por  $ h=\frac{-x^2}{100}+x+100$. Un observador se encuentra parado a 2 pies del fondo del acantilado.
(a) Encuentre la posición del objeto cuando está más cerca del observador.

(b) Encuentre la posición del objeto cuando está más lejos del observador.


**Función objetivo**

Sea el objeto en $(x,h(x))$ con

$$
h(x) \;=\; -\frac{x^2}{100} \;+\; x \;+\;100,
$$

y el observador en el pie del acantilado, a una altura de 2 ft, es decir en $(0,2)$. La distancia euclídea entre ambos es

$$
D(x) \;=\;\sqrt{\bigl(x-0\bigr)^2 \;+\;\bigl(h(x)-2\bigr)^2},
$$

y para simplificar el análisis usamos

$$
Q(x) \;=\; D^2(x)
\;=\;x^2 \;+\;\bigl(h(x)-2\bigr)^2.
$$

El dominio físico va de $x=0$ (lanzamiento) hasta el punto donde $h(x)=0$, i.e.\\

$$
-\tfrac{x^2}{100}+x+100=0
\;\Longrightarrow\;
x = 50\pm 50\sqrt5,
$$

y tomamos la solución positiva
$\;x_{\max} = 50+50\sqrt5\approx161.803$.

\

**Puntos críticos**

* **Fronteras:**

  $$
    Q(0) = 0^2 + (100-2)^2 = 98^2 = 9604,
    \qquad
    Q(50+50\sqrt5) = (50+50\sqrt5)^2 + (0-2)^2 \approx 161.803^2 + 4.
  $$
* **Estacionarios:**

  $$
    Q(x)=x^2 + \Bigl(-\tfrac{x^2}{100}+x+98\Bigr)^2,
  $$

  $$
    Q'(x)
    =2x \;+\;2\bigl(h(x)-2\bigr)\bigl(h'(x)\bigr),
    \quad
    h'(x)=-\tfrac{2x}{100}+1.
  $$

  Al resolver $Q'(x)=0$ en $(0,161.803)$ se obtiene **ninguna raíz real** en el intervalo, de modo que no hay puntos estacionarios interiores.
* **Singularidades:**
  $Q$ es continua y derivable en $[0,161.803]$, salvo en los extremos (donde se evalúa por frontera).

\

**Comparación de valores**

$$
Q(0)=9604,
\qquad
Q(161.803)\approx161.803^2+4\approx26188,
$$

y no hay ningún otro candidato.

\

**Conclusión**

* El **mínimo** de la distancia (el objeto más cerca del observador) se da al **momento del lanzamiento**, en

  $$
    x=0,\quad h(0)=100.
  $$
* El **máximo** de la distancia (el objeto más lejos) se da al **impactar el suelo**, en

  $$
    x=50 + 50\sqrt5\approx161.803,\quad h=0.
  $$

\

**Respuesta final:**

$$
\text{(a) La posición más cercana es }(x,h)=(0,100)
$$

$$
\text{(b) la más lejana es }(x,h)=\bigl(50+50\sqrt5,\;0\bigr).
$$

$$P(x) = 5.00 - 0.002x$$

Encuentre las expresiones para el ingreso, el costo y la utilidad marginales. Determine el nivel de producción que producirá la máxima utilidad total.Explique con claridad cada proceso

Recordando que, ingreso es,

$$ I(P, x) = Px $$

donde,

$ P = \text{Precio} $

$ x = \text{Unidades} $

Teniendo,

$$P(x) = 5.00 - 0.002x$$

Reemplazando $P(x)$ en $I(p, x)$ quedando,

$$ I(x) = x (5.00 - 0.002x) $$

Evaluando,

$$ I(x) = 5.00x - 0.002x^{2} $$

Hallando el precio marginal derivando,

$$ I'(x) = 5.00 - (2)0.002x$$

$$ I'(x) = 5.00 - 0.004x$$

Ahora, el costo que se tiene es,

$$C(x)=3.00 + 1.10x$$

Derivando para hallar el costo marginal,

$$ C'(x) = 1.10 $$

Recordando que Utilidad es,

$$ U = I - C $$

donde,

$ I = \text{Ingreso} $

$ C = \text{Costo} $

$ U = \text{Utilidad} $

Reemplazando las funciones $I(x)$ y $C(x)$,

$$ U(x) = 5.00x - 0.002x^{2} - (3.00 + 1.10x) $$

$$ U(x) = 5.00x - 0.002x^{2} - 3.00 - 1.10x $$

$$ U(x) = -3.00 + 3.90x - 0.002x^{2} $$

Derivando para obtener la Utilidad Marginal,

$$ U'(x) = 3.9 - 0.004x $$

Que para obtener el máximo de utilidades a partir de las unidades producidas, se iguala a 0 la utilidad marginal.

$$ U'(x) = 3.9 - 0.004x = 0$$

$$ U'(x) = 3.9 = 0.004x$$

$$ U'(x) = \frac{3.9}{0.004} = x$$

$$ U'(x) = x = 975 $$

## Método de Newton

````python
 #Analice el siguiente codigo. Explique cada paso en el codigo anterior y compruebe

def newton(fun, der, x_n, epsilon=1e-6, steps=50):
    for n in range(steps + 1):
        # Evaluación de la función para ver si el resultado es válido
        f_x = fun(x_n)
        if abs(f_x) < epsilon:
            return x_n

        # Evaluación de la derivada
        d_f = der(x_n)
        if d_f == 0:
            print('Error la derivada es cero')
            return None

        # Estimación del siguiente punto
        x_n = x_n - f_x / d_f

        return None
````

````python
f=lambda x: x**3-2
g=lambda x: 3*x**2
````

````python
y=newton(f,g,1, epsilon=1e-5)
print(y)
````



**1. Definición de la función**

```python
def newton(fun, der, x_n, epsilon=1e-6, steps=50):
```

* **`fun`**: función $f(x)$ de la que buscamos una raíz.
* **`der`**: derivada $f'(x)$.
* **`x_n`**: conjetura inicial $x_0$.
* **`epsilon`**: tolerancia para decidir que $f(x_n)$ es “suficientemente cero”.
* **`steps`**: número máximo de iteraciones permitidas.

\

**2. Bucle principal de iteraciones**

```python
for n in range(steps + 1):
```

* Recorre $n = 0,1,2,\dots,\text{steps}$.
* Cada iteración aplica un paso de Newton–Raphson.

\

**3. Evaluación de la función**

```python
    f_x = fun(x_n)
    if abs(f_x) < epsilon:
        return x_n
```

* Calcula $f(x_n)$.
* Si $\lvert f(x_n)\rvert < \epsilon$, considera que la raíz ya está aproximada y devuelve $x_n$.

\


**4. Evaluación de la derivada**

```python
    d_f = der(x_n)
    if d_f == 0:
        print('Error la derivada es cero')
        return None
```

* Calcula $f'(x_n)$.
* Si $f'(x_n)=0$, la tangente es horizontal y no puede proseguir; se imprime un mensaje de error y se aborta devolviendo `None`.

\


**5. Cálculo de la siguiente iteración**

```python
    x_n = x_n - f_x / d_f
```

* Aplica la fórmula de Newton:

$$
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}.
$$

\

**6.Error de indentación(retorno prematuro)**

```python
    return None
```

* Como está ubicado, este `return None` se ejecuta **en la primera iteración**, justo después del cálculo de `x_n`.
* Como consecuencia, la función **nunca** hace más de un ciclo: si $f(x_0)$ no es ya menor que $\epsilon$, sale inmediatamente devolviendo `None`.
* **Corrección**: ese `return None` debería quedar **fuera** del bucle `for`, para que solo se active si se agotan las iteraciones sin converger.

\

**7. Ejemplo de uso y comprobación**

```python
f = lambda x: x**3 - 2
g = lambda x: 3*x**2

y = newton(f, g, 1, epsilon=1e-5)
print(y)
```

* Buscamos $\sqrt[3]{2}$: la raíz real de $x^3-2=0$.
* Con $x_0=1$ y $\epsilon=10^{-5}$:

  1. **Iteración 0**: $f(1)=1^3-2=-1$, $|-1|>\epsilon$.
  2. $f'(1)=3$,

     $$
     x_1 = 1 - \frac{-1}{3} = 1 + \tfrac13 = 1.3333\ldots
     $$
  3. Aquí el código retorna `None` y se detiene


— la salida es `None`, no la raíz. Esto confirma el **retorno prematuro** dentro del bucle.

**8. Versión corregida**

Para que el algoritmo funcione correctamente, hay que desplazar el `return None` fuera del `for`, así:

````python
def newton_new(fun, der, x_n, epsilon=1e-6, steps=50):
    for n in range(steps + 1):
        f_x = fun(x_n)
        if abs(f_x) < epsilon:
            return x_n
        d_f = der(x_n)
        if d_f == 0:
            print('Error la derivada es cero')
            return None
        x_n = x_n - f_x / d_f

    # Si llega aquí, no convergió en 'steps' iteraciones
    return None
````

````python
f=lambda x: x**3-2
g=lambda x: 3*x**2
````

````python
y=newton_new(f,g,1, epsilon=1e-5)
print(y)
````


Con esta corrección, el método realiza hasta 50 iteraciones y **devuelve la aproximación** deseada cuando alcanza la tolerancia o `None` si no converge.

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
def newton_new(fun, der, x_n, epsilon=1e-4, steps=50):
    for _ in range(steps + 1):
        f_x = fun(x_n)
        if abs(f_x) < epsilon:
            return x_n
        d_f = der(x_n)
        if d_f == 0:
            return None
        x_n = x_n - f_x / d_f
    return None
````

````python
def find_roots(fun, der, guesses, epsilon=1e-4, tol=1e-4):
    roots = []
    for x0 in guesses:
        r = newton_new(fun, der, x0, epsilon)
        if r is not None:
            root = float(round(r, 6))
            if all(abs(root - rr) > tol for rr in roots):
                roots.append(root)
    return roots
````

 1. $f(x)=\frac{x}{x+1}$

 $  \text{Dominio:}$

 $ x + 1 \neq 0$

 $ x \neq -1$

 $ \mathbb{R}- \begin{Bmatrix}
-1 \end{Bmatrix} $

  $  \text{Rango:}$

  $  \text{Se halla la función inversa}$

  $ y = \frac{x}{x+1} $

  $  \text{Cambiando y por x y despejando y,}$

  $ x = \frac{y}{y+1} $

  $ x (y+1) = y $

  $ xy + x = y $

  $ x = y - xy $

  $ x = y (1 - x) $

  $  y = \frac{x}{1-x} $

  $ 1 - x \neq 0$

  $ x \neq 1$

  $ \mathbb{R}- \begin{Bmatrix} 1 \end{Bmatrix} $

  $ \text{Si } f(-x) = f(x) \text{ Entonces es función par }$

  $ \text{Si } f(-x) = -f(x) \text{ Entonces es función impar }$

  $ \text{Evaluando, } $

  $ f(-x)=\frac{-x}{-x+1} $
  
  $ f(-x)=\frac{-x}{-(x-1)} $

  $ f(-x)=\frac{x}{x-1} $

  $ \text{Por tanto, no es par } $

  $ \text{Evaluando, } $

  $ -f(x)=\frac{-x}{x+1} $
  
  $ \text{Por tanto, no es impar } $

  $ \text{Intersecciones } $

  $ x = 0 $

  $ f(0)=\frac{0}{0+1} = 0 $

  $f'(x)=\frac{(1) (x+1) - x (1)}{(x+1)^{2}}$

  $f'(x)=\frac{x+1 - x}{(x+1)^{2}}$

  $f'(x)=\frac{1}{(x+1)^{2}}$

  $ \text{No tiene puntos críticos. } $

  $f''(x)=\frac{(0)(x+1) - 1 (1)}{(x+1)^{4}}$

  $f''(x)=\frac{- 1}{(x+1)^{4}}$

  $ \text{No es concáva hacia arriba ni hacia abajo } $

  $ x = -1 \text{ -> Asíntota Vertical} $

  $f(x)=\frac{x}{x+1} \text{ Tienen mismo grado en el numerador y denominador. Por tanto se dividen los coeficientes de las variables con grado mayor.}$

  $ \frac{1}{1} = 1 = y $

  $ y = 1 \text{ -> Asíntota Horizontal} $

 3. $f(x)=\frac{x^2}{x^2-1}$

 $  \text{Dominio:}$

 $ (x + 1)(x-1) \neq 0$

 $ x \neq -1$

 $ x \neq 1$

 $ \mathbb{R}- \begin{Bmatrix}
-1, 1 \end{Bmatrix} $

  $  \text{Rango:}$

  $  \text{Se halla la función inversa}$

  $ y = \frac{x^2}{x^2-1} $

  $  \text{Cambiando y por x y despejando y,}$

  $ x = \frac{y^{2}}{y^{2}+1} $

  $ x (y^{2}+1) = y^{2} $

  $ xy^{2} + x = y^{2} $

  $ x = y^{2} - xy^{2} $

  $ x = y^{2} (1 - x) $

  $  y^{2} = \frac{x}{1-x} $

  $  y = \sqrt\frac{x}{1-x} $

  $ 1 - x \neq 0$

  $ x \neq 1$

  $ \frac{x}{1-x} \geq  0$

  $ [-\infty, 1] $

  $ \text{Si } f(-x) = f(x) \text{ Entonces es función par }$

  $ \text{Si } f(-x) = -f(x) \text{ Entonces es función impar }$

  $ \text{Evaluando, } $

  $ f(-x)=\frac{(-x)^{2}}{(-x)^{2}+1} $
  
  $ f(-x)=\frac{x^{2}}{x^{2} - 1} $

  $ \text{Por tanto, es par } $

  $ \text{Intersecciones } $

  $ x = 0 $

  $ f(0) = \frac{0^{2}}{0^{2}-1} = 0

  $f'(x)=\frac{(2x) (x^{2}-1) - x^{2} (2x)}{(x^{2}-1)^{2}}$

  $f'(x)=\frac{2x^{3}-2x-2x^{3}}{(x^{2}-1)^{2}}$

  $f'(x)=\frac{-2x}{(x^{2}-1)^{2}}$

  $ \text{Tiene puntos críticos. } $

  $f'(x)=\frac{-2x}{(x^{2}-1)^{2}} = 0$

  $ x = 0 $

  $ \text{Es creciente para } x < 0$

  $ \text{Es decreciente para } x > 0$

  $ \text{Es un máximo porque primero es creciente y luego decreciente } $

  $f''(x)=\frac{-2(x^{2} - 1)^{2} - (-2x)2(x^{2} - 1) (2x)}{(x^{2}-1)^{4}}$

  $f''(x)=\frac{-2(x^{2} - 1)^{2} + 8x ^{2} (x^{2} - 1)}{(x^{2}-1)^{4}}$

  $f''(x) = \frac{2(3x^{2}+1)}{(x^{2}-1)^{3}}$

  $ \text{No tiene puntos de inflexión } $

  $ x = -1 \text{ -> Asíntota Vertical} $

  $ x = 1 \text{ -> Asíntota Vertical} $

  $f(x)=\frac{x^{2}}{x^{2}-1} \text{ Tienen mismo grado en el numerador y denominador. Por tanto se dividen los coeficientes de las variables con grado mayor.}$

  $ \frac{1}{1} = 1 = y $

  $ y = 1 \text{ -> Asíntota Horizontal} $
